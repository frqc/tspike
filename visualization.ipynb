```python
import numpy as np 
import torch
from tnn import *
import matplotlib.pyplot as plt
```

```python
state = torch.load('C:\\Users\\nvagus\\Documents\\CMU\\18847\\tspike\\model\\n-mnist-cv')
weight = state['weight']
fig, axes = plt.subplots(8, 2, figsize=(10, 7))

for i, w in enumerate(weight.cpu()):
    for j in range(2):
        ax = axes[i][j]
        ax.imshow(w[j].T)
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
```
